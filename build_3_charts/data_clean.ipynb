```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('raw_data.json')
```

```python
df = df.dropna()
```

```python
histogram_df = df.groupby(['Age']).count().reset_index()
```

```python
Ages = histogram_df['Age']
Counts = histogram_df['FIELD1']
```

```python
output_histogram = pd.DataFrame()
output_histogram['age'] = Ages
output_histogram['num'] = Counts
```

```python
output_histogram
```

```python
output_histogram.to_csv('age_of_shooters.csv', index=False)
```

```python
import datetime as dt
df.head(3)
```

```python
df['date'] = pd.to_datetime(df['Date'],unit='s')
```

```python
df['month'] = df['date'].dt.month
```

```python
df.head(3)
```

```python
scatter_df = df.groupby(['month']).count().reset_index()
```

```python
month = scatter_df['month']
count = scatter_df['FIELD1']
```

```python
scatter_df = pd.DataFrame()
scatter_df['month'] = month
scatter_df['count'] = count
```

```python
scatter_df['month'] = scatter_df['month'].apply(lambda x: str(x) + '/' + '15')
```

```python
scatter_df.head(12)
```

```python
scatter_df.to_csv('month_of_shootings.csv',index=False)
```
